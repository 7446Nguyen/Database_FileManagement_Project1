```python
import pandas as pd
import seaborn as sea
import matplotlib.font_manager
```

```python
performance = pd.read_csv("/media/jeff/dataSci/MSDS/2020 Spring/DB/Project/Performance_Summary.csv")
```

```python
performance.head(10)
```

```python
extraction = performance[performance['Sample'].str.contains('Extract', regex=True)]
transform = performance[performance['Sample'].str.contains('Trans', regex=True)]
load = performance[performance['Sample'].str.contains('Load', regex=True)]

extraction['Sample'].value_counts()
transform['Sample'].value_counts()
load['Sample'].value_counts()
```

```python
# Set plot styles
sea.set(font = "Times New Roman")
import matplotlib as mpl 

sea.set_style("white")
sea.set_style({'font.family': 'Times New Roman'})
```

```python


order=["lightExtract", "lightExtractMongo", "medExtract","medExtractMongo","heavyExtract","heavyExtractMongo"]
extraction_performance = sea.boxplot(y = "Sample", x = "elapsed_time", data = extraction, order = order)
extraction_performance.set(xlabel="Elapsed time (sec)", ylabel='Extraction Type')
```

```python
extraction_performance = sea.boxplot(y = "Sample", x = "CPU_d", data = extraction, order = order)
extraction_performance.set(xlabel="CPU Utilization - Difference from Base Reading (%)", ylabel='Extraction Type')
```

```python
extraction_performance = sea.boxplot(y = "Sample", x = "RAM_d", data = extraction, order = order)
extraction_performance.set(xlabel="RAM Utilization - Difference from Base Reading (%)", ylabel='Extraction Type')
```
